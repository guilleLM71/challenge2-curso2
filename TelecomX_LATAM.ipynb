```python
# Ahora que los datos est√°n limpios, es momento de crear la columna "Cuentas_Diarias". Utiliza la facturaci√≥n mensual para calcular el valor diario, proporcionando una visi√≥n m√°s detallada del comportamiento de los clientes a lo largo del tiempo.

import pandas as pd
# Convertir 'TotalCharges' a num√©rico, manejando errores convirtiendo a NaN
telecom['Charges.Total'] = pd.to_numeric(telecom['Charges.Total'], errors='coerce')

# Llenar valores NaN en 'TotalCharges' con 0 o la mediana, dependiendo del contexto.
# Si la ausencia indica 0 cargos, llenar con 0. Si es un valor faltante que deber√≠a existir, usar la mediana.
# Asumiendo que NaN en TotalCharges indica 0 cargos al inicio del servicio:
telecom['Charges.Total'] = telecom['Charges.Total'].fillna(0)

# Crear la columna 'Cuentas_Diarias'
# Se calcula la facturaci√≥n diaria dividiendo 'MonthlyCharges' por 30 (un promedio de d√≠as en un mes).
telecom['Cuentas_Diarias'] = telecom['Charges.Monthly'] / 30

# Mostrar las primeras filas con la nueva columna
print(telecom[['Charges.Monthly', 'Cuentas_Diarias']].head())

# Verificar el tipo de datos de la nueva columna
print(telecom.info())
```

```python
# La estandarizaci√≥n y transformaci√≥n de datos es una etapa opcional, pero altamente recomendada, ya que busca hacer que la informaci√≥n sea m√°s consistente, comprensible y adecuada para el an√°lisis. Durante esta fase, por ejemplo, puedes convertir valores textuales como "S√≠" y "No" en valores binarios (1 y 0), lo que facilita el procesamiento matem√°tico y la aplicaci√≥n de modelos anal√≠ticos.

import pandas as pd
# Transformar las columnas binarias 'Yes'/'No' a 1/0
binary_cols = [
    'Churn',
    'Partner',
    'Dependents',
    'PaperlessBilling',
    'OnlineSecurity',
    'OnlineBackup',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies'
]

for col in binary_cols:
    # Se asume que 'Yes' corresponde a 1 y 'No' a 0. Ajusta si es diferente.
    if col in telecom.columns:
      telecom[col] = telecom[col].map({'Yes': 1, 'No': 0, 'No internet service': 0, 'No phone service': 0})


# Transformar la columna 'gender' a binario
# Se asume que 'Female' corresponde a 1 y 'Male' a 0. Ajusta si es diferente.
if 'gender' in telecom.columns:
  telecom['gender'] = telecom['gender'].map({'Female': 1, 'Male': 0})

# Transformar 'PhoneService' a binario
if 'PhoneService' in telecom.columns:
  telecom['PhoneService'] = telecom['PhoneService'].map({'Yes': 1, 'No': 0})

# Transformar 'MultipleLines' a binario, manejando el valor 'No phone service'
if 'MultipleLines' in telecom.columns:
  telecom['MultipleLines'] = telecom['MultipleLines'].map({'Yes': 1, 'No': 0, 'No phone service': 0})


# Aplicar One-Hot Encoding a columnas categ√≥ricas con m√°s de dos categor√≠as
categorical_cols = [
    'InternetService',
    'Contract',
    'PaymentMethod'
]

for col in categorical_cols:
    if col in telecom.columns:
        telecom = pd.get_dummies(telecom, columns=[col], prefix=col, dummy_na=False)

# Mostrar las primeras filas despu√©s de la transformaci√≥n
print("\nDataFrame despu√©s de la estandarizaci√≥n y transformaci√≥n:")
print(telecom.head())

# Verificar los tipos de datos despu√©s de la transformaci√≥n
print("\nTipos de datos despu√©s de la transformaci√≥n:")
print(telecom.info())
```

```python
# Para comenzar, realiza un an√°lisis descriptivo de los datos, calculando m√©tricas como media, mediana, desviaci√≥n est√°ndar y otras medidas que ayuden a comprender mejor la distribuci√≥n y el comportamiento de los clientes.

# An√°lisis descriptivo de las columnas num√©ricas
print("\nAn√°lisis descriptivo de las columnas num√©ricas:")
print(telecom.describe())

# Para columnas categ√≥ricas, podemos ver la frecuencia de cada categor√≠a
print("\nAn√°lisis descriptivo de las columnas categ√≥ricas:")
for col in telecom.columns:
    if telecom[col].dtype == 'object' or (telecom[col].dtype == 'int64' and col in relevant_columns and col not in ['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges', 'Cuentas_Diarias']): # Excluir columnas num√©ricas ya descritas y binarias transformadas
        print(f"\nFrecuencia de categor√≠as en '{col}':")
        print(telecom[col].value_counts())

# Tambi√©n podemos calcular la mediana y la desviaci√≥n est√°ndar para las columnas num√©ricas si describe() no las incluye expl√≠citamente por alg√∫n motivo
# (describe() ya las incluye, pero esto es para demostrar c√≥mo hacerlo si fuera necesario)
print("\nMediana de columnas num√©ricas:")
print(telecom[['tenure', 'Charges.Monthly', 'Charges.Total', 'Cuentas_Diarias']].median())

print("\nDesviaci√≥n est√°ndar de columnas num√©ricas:")
print(telecom[['tenure', 'Charges.Monthly', 'Charges.Total', 'Cuentas_Diarias']].std())
```

```python
# En este paso, el objetivo es comprender c√≥mo est√° distribuida la variable "churn" (evasi√≥n) entre los clientes. Utiliza gr√°ficos para visualizar la proporci√≥n de clientes que permanecieron y los que se dieron de baja.

import matplotlib.pyplot as plt
import seaborn as sns

# Visualizar la distribuci√≥n de la variable 'Churn'
plt.figure(figsize=(6, 4))
sns.countplot(x='Churn', data=telecom)
plt.title('Distribuci√≥n de Clientes por Evasi√≥n (Churn)')
plt.xlabel('Evasi√≥n (Churn)')
plt.ylabel('N√∫mero de Clientes')
plt.xticks(ticks=[0, 1], labels=['No (Permanecen)', 'S√≠ (Se dan de baja)'])
plt.show()

# Visualizar la proporci√≥n de clientes que permanecieron y los que se dieron de baja usando un gr√°fico de pastel
churn_counts = telecom['Churn'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(churn_counts, labels=['No (Permanecen)', 'S√≠ (Se dan de baja)'], autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Proporci√≥n de Clientes que Permanecen vs. Clientes que se dan de Baja')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Ahora, exploraremos c√≥mo se distribuye la evasi√≥n seg√∫n variables categ√≥ricas, como g√©nero, tipo de contrato, m√©todo de pago, entre otras.
# Este an√°lisis puede revelar patrones interesantes, por ejemplo, si los clientes de ciertos perfiles tienen una mayor tendencia a cancelar el servicio, lo que ayudar√° a orientar acciones estrat√©gicas.

import matplotlib.pyplot as plt
# Analizar la distribuci√≥n de la evasi√≥n por variables categ√≥ricas
categorical_vars_to_analyze = [
    'gender',
    'SeniorCitizen',
    'Partner',
    'Dependents',
    'PhoneService',
    'MultipleLines',
    'OnlineSecurity',
    'OnlineBackup',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies',
    'PaperlessBilling',
    'InternetService_DSL', # Columnas one-hot encoded
    'InternetService_Fiber optic',
    'InternetService_No',
    'Contract_Month-to-month',
    'Contract_One year',
    'Contract_Two year',
    'PaymentMethod_Bank transfer (automatic)',
    'PaymentMethod_Credit card (automatic)',
    'PaymentMethod_Electronic check',
    'PaymentMethod_Mailed check'
]

for col in categorical_vars_to_analyze:
    if col in telecom.columns:
        plt.figure(figsize=(8, 5))
        sns.countplot(x=col, hue='Churn', data=telecom)
        plt.title(f'Distribuci√≥n de Evasi√≥n por {col}')
        plt.xlabel(col)
        plt.ylabel('N√∫mero de Clientes')
        plt.xticks(rotation=0) # Ajusta la rotaci√≥n si las etiquetas se superponen
        plt.legend(title='Evasi√≥n', labels=['No', 'S√≠'])
        plt.show()

        # Opcional: Mostrar la proporci√≥n de evasi√≥n por categor√≠a en forma de tabla
        print(f"\nProporci√≥n de Evasi√≥n por {col}:")
        churn_proportion = telecom.groupby(col)['Churn'].value_counts(normalize=True).unstack()
        print(churn_proportion)
        print("-" * 30)
```

```python
# En este paso, explora c√≥mo las variables num√©ricas, como "total gastado" o "tiempo de contrato", se distribuyen entre los clientes que cancelaron (evasi√≥n) y los que no cancelaron.
# Este an√°lisis ayuda a entender si ciertos valores num√©ricos est√°n m√°s asociados con la evasi√≥n, proporcionando insights sobre los factores que influyen en el comportamiento de los clientes.

import matplotlib.pyplot as plt
# An√°lisis de la distribuci√≥n de variables num√©ricas entre clientes que cancelaron y no cancelaron

numeric_vars_to_analyze = [
    'tenure',          # Tiempo de contrato en meses
    'Charges.Monthly', # Cargo mensual
    'Charges.Total',   # Cargo total
    'Cuentas_Diarias'  # Cargo diario (calculado)
]

# Graficar la distribuci√≥n de cada variable num√©rica por 'Churn'
for col in numeric_vars_to_analyze:
    if col in telecom.columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(data=telecom, x=col, hue='Churn', multiple="stack", kde=True)
        plt.title(f'Distribuci√≥n de {col} por Evasi√≥n (Churn)')
        plt.xlabel(col)
        plt.ylabel('Frecuencia')
        plt.show()

        # Opcional: Mostrar estad√≠sticas descriptivas para cada variable num√©rica por 'Churn'
        print(f"\nEstad√≠sticas descriptivas de '{col}' por Evasi√≥n:")
        print(telecom.groupby('Churn')[col].describe())
        print("-" * 30)

# Tambi√©n puedes usar boxplots para visualizar la distribuci√≥n y comparar medianas
for col in numeric_vars_to_analyze:
    if col in telecom.columns:
        plt.figure(figsize=(10, 6))
        sns.boxplot(data=telecom, x='Churn', y=col)
        plt.title(f'Boxplot de {col} por Evasi√≥n (Churn)')
        plt.xlabel('Evasi√≥n (Churn)')
        plt.ylabel(col)
        plt.xticks(ticks=[0, 1], labels=['No', 'S√≠'])
        plt.show()
        print("-" * 30)
```

El an√°lisis se centrar√° en identificar patrones, factores y caracter√≠sticas de los clientes que est√°n asociados a una mayor probabilidad de darse de baja del servicio, con el fin de proporcionar informaci√≥n valiosa para la toma de decisiones estrat√©gicas y la implementaci√≥n de medidas de retenci√≥n efectivas.
